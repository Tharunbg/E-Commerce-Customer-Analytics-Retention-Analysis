# 🚀 STEP 2 - Data Cleaning and Master Table Creation

Goal:

 STEP 2.1 — Filter Valid Orders

```python
orders['order_status'].value_counts()
```

```python
orders_clean = orders[orders['order_status'] == 'delivered']
```

```python
print(orders.shape)
print(orders_clean.shape)
```

 STEP 2.2 — Fix Date Columns

```python
orders_clean['order_purchase_timestamp'] = pd.to_datetime(
    orders_clean['order_purchase_timestamp']
)
```

```python
orders_clean['order_date'] = orders_clean['order_purchase_timestamp'].dt.date
```

 STEP 2.3 — Handle Payments

```python
order_payments.groupby('order_id').size().value_counts()
```

```python
payments_agg = order_payments.groupby('order_id').agg({
    'payment_value': 'sum'
}).reset_index()
```

 STEP 2.4 — Merge Customers (CRITICAL STEP)

Use customer_unique_id (NOT customer_id)

```python
orders_customers = orders_clean.merge(
    customers[['customer_id', 'customer_unique_id', 'customer_city', 'customer_state']],
    on='customer_id',
    how='left'
)
```

 STEP 2.5 — Merge Order Items

```python
orders_items = orders_customers.merge(
    order_items,
    on='order_id',
    how='left'
)
```

 STEP 2.6 — Merge Products

```python
orders_products = orders_items.merge(
    products[['product_id', 'product_category_name']],
    on='product_id',
    how='left'
)
```

 STEP 2.7 — Merge Payments

```python
master_df = orders_products.merge(
    payments_agg,
    on='order_id',
    how='left'
)
```

Now we have MASTER DATASET

# COHORT IN PYTHON ->customer_orders dataset

```python
import pandas as pd

cohort_df = pd.read_csv('/content/sample_data/customer_orders.csv')

cohort_df.head()
```

```python
#STEP 2 — Convert to datetime
cohort_df['order_date'] = pd.to_datetime(cohort_df['order_date'])
```

```python
#STEP 3 — Get first purchase per customer
cohort_df['cohort_month'] = cohort_df.groupby('customer_unique_id')['order_date'].transform('min')
```

```python
#STEP 4 — Convert to month format
cohort_df['cohort_month'] = cohort_df['cohort_month'].dt.to_period('M')
cohort_df['order_month'] = cohort_df['order_date'].dt.to_period('M')
```

```python
#STEP 5 — Calculate month index
cohort_df['month_number'] = (
    (cohort_df['order_month'] - cohort_df['cohort_month'])
).apply(lambda x: x.n)
```

```python
#STEP 6 Group data
cohort_data = cohort_df.groupby(
    ['cohort_month', 'month_number']
)['customer_unique_id'].nunique().reset_index()
```

```python
#STEP 7 — Pivot table
cohort_pivot = cohort_data.pivot(
    index='cohort_month',
    columns='month_number',
    values='customer_unique_id'
)
```

```python
#STEP 8 — Convert to retention %
cohort_size = cohort_pivot.iloc[:, 0]

retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
```

```python
#STEP 9 — HEATMAP
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

sns.heatmap(
    retention_matrix,
    annot=True,
    fmt=".0%",
    cmap="coolwarm"
)

plt.title("Customer Retention Cohort Analysis")
plt.show()
```

# STEP 6 — RFM SEGMENTATION

What is RFM?\
Recency-> How recently customer purchased\
Frequency->	How often customer purchases\
Monetary->	How much customer spends

___
🎯 Goal

we will:


*   Score customers
*   Segment them
*   Identify:
High-value customers 💰\
At-risk customers ⚠️\
Lost customers ❌

___
We use:

👉 customer_orders (for Recency + Frequency)\
👉 master_data (for Monetary)

STEP 6.1 — LOAD DATA

```python
import pandas as pd

orders_df = pd.read_csv('/content/olist_orders_dataset.csv')
master_df = pd.read_csv('/content/cleaned_dataset.csv')
```

STEP 6.2 — PREPARE DATES

```python
orders_df['order_purchase_timestamp'] = pd.to_datetime(
    orders_df['order_purchase_timestamp']
)

orders_df['order_date'] = orders_df['order_purchase_timestamp'].dt.date
```

```python
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])
master_df['order_date'] = pd.to_datetime(master_df['order_date'])
```

STEP 6.3 — CALCULATE RECENCY

👉 Recency = days since last purchase

```python
snapshot_date = orders_df['order_date'].max() + pd.Timedelta(days=1)

recency = orders_df.groupby('customer_id')['order_date'].max().reset_index()
recency['recency'] = (snapshot_date - recency['order_date']).dt.days

recency = recency[['customer_id', 'recency']]
```

STEP 6.4 — CALCULATE FREQUENCY

```python
frequency = orders_df.groupby('customer_id')['order_id'].nunique().reset_index()
frequency.columns = ['customer_id', 'frequency']
```

STEP 6.5 — CALCULATE MONETARY

```python
monetary = master_df.groupby('customer_id')['revenue'].sum().reset_index()
monetary.columns = ['customer_id', 'monetary']
```

STEP 6.6 — MERGE RFM

```python
rfm = recency.merge(frequency, on='customer_id')
rfm = rfm.merge(monetary, on='customer_id')

rfm.head()
```

**STEP 6.7 — CREATE SCORES**

Recency Score (LOW is better)

```python
rfm['R_score'] = pd.qcut(rfm['recency'], 5, labels=[5,4,3,2,1])
```

Frequency Score (HIGH is better)

```python
rfm['F_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
```

Monetary Score (HIGH is better)

```python
rfm['M_score'] = pd.qcut(rfm['monetary'], 5, labels=[1,2,3,4,5])
```

STEP 6.8 — COMBINE RFM SCORE

```python
rfm['RFM_score'] = (
    rfm['R_score'].astype(str) +
    rfm['F_score'].astype(str) +
    rfm['M_score'].astype(str)
)
```

STEP 6.9 — CREATE CUSTOMER SEGMENTS

```python
def segment_customer(row):
    if row['R_score'] == 5 and row['F_score'] >= 4:
        return 'Champions'
    elif row['R_score'] >= 4 and row['F_score'] >= 3:
        return 'Loyal Customers'
    elif row['R_score'] <= 2 and row['F_score'] <= 2:
        return 'At Risk'
    else:
        return 'Others'

rfm['segment'] = rfm.apply(segment_customer, axis=1)
```

STEP 6.10 — ANALYZE SEGMENTS

```python
rfm['segment'].value_counts()
```

Revenue by segment

```python
rfm_segment_revenue = rfm.merge(
    master_df[['customer_id', 'revenue']],
    on='customer_id'
)

rfm_segment_revenue.groupby('segment')['revenue'].sum()
```

STEP 6.11 — VISUALIZE

```python
import matplotlib.pyplot as plt

rfm['segment'].value_counts().plot(kind='bar')
plt.title("Customer Segments Distribution")
plt.show()
```

The business is at high risk of losing its most valuable customers.

Found that a large portion of revenue comes from customers classified as ‘At Risk’, which indicates a significant risk of revenue loss if these customers are not retained.

Insight 1

A majority of customers fall into low-engagement segments, indicating weak repeat purchase behavior.

Insight 2

A significant portion of revenue is generated by customers classified as “At Risk,” highlighting a major retention challenge.

Insight 3

The number of high-value “Champion” customers is relatively small, suggesting an opportunity to grow this segment.
